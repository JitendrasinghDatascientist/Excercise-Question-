```python
def calculate_average_salary_increase(starting_salary, num_years, yearly_increase_percentage):

    salaries = []
 
    current_salary = starting_salary
    for year in range(num_years):
        salaries.append(current_salary)
        current_salary *= (1 + (yearly_increase_percentage / 100))
    

    total_salary_increase = sum(salaries) - (starting_salary * num_years)
    average_salary_increase = total_salary_increase / (num_years - 1)  
    
    return average_salary_increase

starting_salary = 50000  
num_years = 5            
yearly_increase_percentage = 3  

average_increase = calculate_average_salary_increase(starting_salary, num_years, yearly_increase_percentage)
print("Year-wise average salary increase:", average_increase)
```

```python
def Calculate_year_wise_avg_sal(Starting_salary,num_years,yearly_increasing_salary):
    salary = []
    current_sal = Starting_salary
    for year in range(num_year):
        salary.append(current_sal)
        current_sal *= (1+(yearly_increasing_salary/100))
        
    total_sal_increase = sum(salary)-(Starting_salary * num_year)
    avg_year_sal = total_sal_increase/(num_year - 1)
    return avg_year_sal
starting_salary = 50000
num_year = 5
yearly_increasing_salary = 3
avg_increase = Calculate_year_wise_avg_sal(starting_salary,num_year,yearly_increasing_salary)
print(avg_increase)
        
```
